```python
import numpy as np
```

```python
nums = list(numpy.arange(0,301))
```

```python
nums_dict = {}
```

```python
count = [0, 1, 2]*100
```

```python
for a in count:
    nums_dict[nums[0]] = a
    my_count = 0
    nums.pop(my_count)
    my_count + 1
```

```python
nums_dict
```

```python
data0 = [ k for k, v in nums_dict.items() if v == 0 ]
data1 = [ k for k, v in nums_dict.items() if v == 1 ]
data2 = [ k for k, v in nums_dict.items() if v == 2 ]
```

```python
for idx, x in nums_dict.items():
    print(idx, x)
```

```python
def check_value(val):
    if val in data0:
        print("data0")
    elif val in data1:
        print("data1")
    elif val in data2:
        print("data2")
    else: print("not listed")
```

```python
check_value(2)
```
